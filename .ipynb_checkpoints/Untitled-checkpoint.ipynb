```python
#  Read the three files into python as dataframes

df1 = pd.read_csv("/Users/elnazshahzeidi/Downloads/file1.csv")
df2 = pd.read_csv("/Users/elnazshahzeidi/Downloads/file2.csv")
df3 = pd.read_csv("/Users/elnazshahzeidi/Downloads/file3.csv")



data = pd.concat([df1,df2,df3]).reset_index(drop=True)
data
```

```python
data = df1._append(df2)._append(df3).reset_index(drop=True)
data
```

```python
#  Show the DataFrame's shape.

data.shape
```

```python
#  Standardize header names.
    
data.columns = data.columns.str.lower().str.replace(' ', '_')
data.columns 
   

```

```python
# Which columns are numerical?

data.dtypes
```

```python
numerical_columns = data.select_dtypes(include=['number']).columns.tolist()
numerical_columns
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
categorical_columns 
```

```python
# Understand the meaning of all columns
print(data.head())
```

```python
# Delete the column education and the number of open complaints from the dataframe.


data = data.drop(columns=["education","number_of_open_complaints"])
data.head()


```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

data['customer_lifetime_value'] = data['customer_lifetime_value'].str.rstrip('%').astype(float) / 100.0

data['customer_lifetime_value']
```

```python
# Check for duplicate rows in the data and remove if any.

data.drop_duplicates()
```

```python
# Filter out the data for customers who have an income of 0 or less.

data = data[data['income'] > 0]
    
data
```
